# Most and Least Expensive Choices

```python
raw_Data.groupby(['LPG SIZE']).max()
```

```python
raw_Data.groupby(['LPG SIZE']).min()
```

```python
group_Brands_Size = raw_Data.groupby(['LPG BRANDS', 'LPG SIZE'])['LATEST PRICE'].mean().reset_index();

group_Brands_Size.head(20)
```

```python
gbc_Brands_Size = px.bar(group_Brands_Size, x="LPG SIZE", y='LATEST PRICE', color="LPG BRANDS", barmode="group")
gbc_Brands_Size.show()
```

```python
group_City_Brand = raw_Data.groupby(['CITY/ MUNICIPALITY', 'LPG BRANDS'])['LATEST PRICE'].mean().reset_index()

group_City_Brand.head(30)
```

```python
gbc_City_Brands = px.bar(group_City_Brand, x='CITY/ MUNICIPALITY', y='LATEST PRICE', color='LPG BRANDS', barmode='group', title='Error in Data Grouping')
gbc_City_Brands.show()
```
